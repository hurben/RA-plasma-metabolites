```r
calculate_significance <- function(x, m){

#number of sig metabolites in pathway
# x <- 3
#number of metabolites in pathway
# m <- 7
#number of metabolites in pathway
n <- 686
#significantly changed metabolites
k <- 33

data_matrix <- matrix(c(x, k - x, m - x, n - k - m + x), 2, 2)
colnames(data_matrix) <- c("Sig","notSig")
rownames(data_matrix) <- c("inPathway","notPathway")
    
# print (data_matrix)
hyper_pval <- phyper(x - 1, k, n - k, m, lower.tail=FALSE)
fet_pval <- fisher.test(data_matrix)$p.value
                     
hyper_pval <- sprintf(hyper_pval, fmt = '%#.4f')
fet_pval <- sprintf(fet_pval, fmt = '%#.4f')                
                     
if (fet_pval < 0.05){
    print (paste0(">> hypergeomteric: ", hyper_pval))
    print (paste0(">> fisher exact test ", fet_pval))
#     print (phyper(x - 1, k, n - k, m, lower.tail=FALSE))
    print ("----------------------------")
}
else{
    print ("not significant")
    print ("----------------------------")
    
}

return (fet_pval)
}
```

```r
data_table <- "/Users/m221138/RA_plasma_metabolites/analysis/pathway_analysis/pathway_data_table_v2.csv"
data_table <- read.csv(data_table)
# data_table
```

```r
NROW <- nrow(data_table)
count <- 0

for (i in 1:NROW){

    print (as.character(data_table[i, 1]))
    
    m <- data_table[i, 2] #number of non-significant metabolites in pathway
    x <- data_table[i, 3] #number of significant metabolites in pathway
    
    fet_pval <- calculate_significance(x, m)
    
    if (fet_pval < 0.05){
        count <- count + 1
    }
    
#     break
}

print ("SUMMARY")
print (paste0("number of significant sub-pathways: ", count))
```
