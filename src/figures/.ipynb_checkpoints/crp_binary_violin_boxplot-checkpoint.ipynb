```python
import pandas as pd
import seaborn as sns
import statistics
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.patches as mpatches
# rc('font',**{'family':'serif','serif':['Palatino']})
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
plt.rcParams['pdf.fonttype'] = 42
```

```python
data_df = pd.read_csv('/Users/m221138/RA_plasma_metabolites/data/67_met_every_confounds_v2.csv', index_col=0)
data_df = data_df.T
data_df
```

```python
met_list = ["803","503","1442","100002945","100001197"]
met_name_list = ['mannose','serine','beta-hydroxyisovalerate',
                 '(14 or 15)-methylpalmitate (a17:0 or i17:0)','10-undecenoate (11:1n1)']
```

```python
for i in range(len(met_list)):
    
    chemID = met_list[i]
    chem_name = met_name_list[i]
    print (chemID)
    print (chem_name)
    vp = sns.violinplot(x="crp", y=chemID, data=data_df)
    vp.set(ylabel=chem_name, xlabel="0 (< 3) 1 (>3)")
    vp.get_figure()
    vp.figure.savefig('/Users/m221138/RA_plasma_metabolites/data/binary_crp/%s.%s.pdf' % (chem_name, "violin"))
    
    vp.get_figure().clf()
    
    bp = sns.boxplot(x="crp", y=chemID, data=data_df)
    bp.set(ylabel=chem_name, xlabel="0 (< 3) 1 (>3)")
    bp.get_figure()
    bp.figure.savefig('/Users/m221138/RA_plasma_metabolites/data/binary_crp/%s.%s.pdf' % (chem_name, "box"))

    bp.get_figure().clf()
    
```
