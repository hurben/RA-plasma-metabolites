```python
def info_to_list (visit_dict, data_df, i):
    
    das = float(data_df.iloc[i,3])
    age = float(data_df.iloc[i,4])
    sex = data_df.iloc[i,5]
    bmi = float(data_df.iloc[i,6])

    smoke = str(data_df.iloc[i,7])

    rf = str(data_df.iloc[i,8])
    ccp = str(data_df.iloc[i,9])
 
    try : visit_dict['das'].append(das)
    except KeyError: visit_dict['das'] = [das]
    
    try : visit_dict['age'].append(age)
    except KeyError: visit_dict['age'] = [age]
        
    try : visit_dict['sex'].append(sex)
    except KeyError: visit_dict['sex'] = [sex]
    
    if math.isnan(bmi) == False:
        try : visit_dict['bmi'].append(bmi)
        except KeyError: visit_dict['bmi'] = [bmi]
            
    try : visit_dict['bmi_full'].append(str(bmi))
    except KeyError: visit_dict['bmi_full'] = [str(bmi)]

    try : visit_dict['smoke'].append(smoke)
    except KeyError: visit_dict['smoke'] = [smoke]
    
    try : visit_dict['rf'].append(rf)
    except KeyError: visit_dict['rf'] = [rf]
        
    try : visit_dict['ccp'].append(ccp)
    except KeyError: visit_dict['ccp'] = [ccp]

    return visit_dict
```
