```python
import pandas as pd
```

```python
data_file = '/Users/m221138/RA_plasma_metabolites/data/DMARD_prescription_profile.csv'
```

```python
data_df = pd.read_csv(data_file)
r, c = data_df.shape
data_df.head()
```

```python
patientID_list =[]
count_dict = {}
print (r)
for i in range(r):
    patientID = data_df.iloc[i,0]
    visit = data_df.iloc[i,1]
    csDMARD_status = data_df.iloc[i,7]
    bDMARD_status = data_df.iloc[i,15]
    pred_status = data_df.iloc[i,16]
    
    if patientID not in patientID_list:
        patientID_list.append(patientID)
    
    if visit == 1:
        if csDMARD_status == 1:
            try: count_dict['csDMARD', 'visit1'] += 1
            except KeyError: count_dict['csDMARD', 'visit1'] = 1
        if bDMARD_status == 1:    
            try: count_dict['bDMARD', 'visit1'] += 1
            except KeyError: count_dict['bDMARD', 'visit1'] = 1
        if pred_status == "yes":
            try: count_dict['pred', 'visit1'] += 1
            except KeyError: count_dict['pred', 'visit1'] = 1
    if visit == 2:
        if csDMARD_status == 1:
            try: count_dict['csDMARD', 'visit2'] += 1
            except KeyError: count_dict['csDMARD', 'visit2'] = 1
        if bDMARD_status == 1:    
            try: count_dict['bDMARD', 'visit2'] += 1
            except KeyError: count_dict['bDMARD', 'visit2'] = 1
        if pred_status == "yes":
            try: count_dict['pred', 'visit2'] += 1
            except KeyError: count_dict['pred', 'visit2'] = 1
```

```python
print (count_dict)
print (len(patientID_list))
```
