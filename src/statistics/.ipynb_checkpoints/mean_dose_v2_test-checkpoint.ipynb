```python
import pandas as pd
import statistics
from scipy.stats import iqr
import numpy
info1_file = '/Users/m221138/RA_plasma_metabolites/data/patient_pred_mtx_info_test.csv'
info2_file = '/Users/m221138/RA_plasma_metabolites/data/pred_mtx_info.csv'
```

```python
info1_df = pd.read_csv(info1_file)
info2_df = pd.read_csv(info2_file)

main_dict = {}
mcID_list = []
dose_dict = {}
```

```python
info1_df
```

```python
info2_df
```

```python
r, c = info1_df.shape
for i in range(r):
    mcID = str(info1_df.iloc[i,1])
    
    if mcID not in mcID_list:
        mcID_list.append(mcID)
    visit = info1_df.iloc[i,2]
    mtx = info1_df.iloc[i,3]
    pred = info1_df.iloc[i,4]
    
    main_dict[mcID, visit,"mtx"] = mtx
    main_dict[mcID, visit,"pred"] = pred
```

```python
main_dict
```

```python
r,c = info2_df.shape
mtx_list = []
pred_list = []

for i in range(r):
    mcID = str(info2_df.iloc[i,0])
    mcID = mcID.replace('.0','')
    
    try: visit = int(info2_df.iloc[i,1])
    except ValueError: visit = 3
     
    mtx_status = info2_df.iloc[i,2]
    mtx_value = info2_df.iloc[i,3]
    pred_status = info2_df.iloc[i,4]
    pred_value = info2_df.iloc[i,5]

    if pred_status == "yes":
        try:
            pred_test_status = main_dict[mcID, visit, 'pred']
            if pred_test_status == "yes":
                print ('1')
                print (pred_value)
                pred_list.append(pred_value)
        except KeyError:
            None

    if mtx_status == "yes":
        try:
            mtx_test_status = main_dict[mcID, visit, 'mtx']
            if mtx_test_status == "yes":
                mtx_list.append(mtx_value) 
        except KeyError:
            None
```

```python
print (len(mtx_list))
print (mtx_list)

print (len(pred_list))
print (pred_list)
```

```python
print ("MTX")
print ("size:", len(mtx_list))
print (statistics.median(mtx_list))
print(iqr(mtx_list, rng=(25,75), interpolation='midpoint'))
print ("1st qt: ",numpy.quantile(mtx_list, .25))
print ("3rd qt: ",numpy.quantile(mtx_list, .75))

print ("PRED")
print ("size:", len(pred_list))
# print (pred1_list)
print (statistics.median(pred_list))
print(iqr(pred_list, rng=(25,75), interpolation='midpoint'))
print ("1st qt: ",numpy.quantile(pred_list, .25))
print ("3rd qt: ",numpy.quantile(pred_list, .75))
```
