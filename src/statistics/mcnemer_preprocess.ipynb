```python
import pandas as pd
```

```python
data_file = '/Users/m221138/RA_plasma_metabolites/data/DMARD_prescription_profile.csv'
```

```python
data_df = pd.read_csv(data_file)
r, c = data_df.shape
data_df.head()
```

```python
patient_info_dict = {}
patientID_list = []
count_dict = {}
print (r)
for i in range(r):
    patientID = data_df.iloc[i,0]
    visit = data_df.iloc[i,1]
    csDMARD_status = data_df.iloc[i,7]
    bDMARD_status = data_df.iloc[i,15]
    pred_status = data_df.iloc[i,16]
    mtx_status = data_df.iloc[i,2]
    
    if patientID not in patientID_list:
        patientID_list.append(patientID)
    
    if visit == 1:
        if csDMARD_status > 0:
            patient_info_dict[patientID, "csDMARD", "visit1"] = 1
        if bDMARD_status > 0:    
            patient_info_dict[patientID, "bDMARD", "visit1"] = 1
        if pred_status == "yes":
            patient_info_dict[patientID, "pred", "visit1"] = 1
        if mtx_status == "yes":
            patient_info_dict[patientID, "mtx", "visit1"] = 1
        
    if visit == 2:
        if csDMARD_status > 0:
            patient_info_dict[patientID, "csDMARD", "visit2"] = 1
        if bDMARD_status > 0:    
            patient_info_dict[patientID, "bDMARD", "visit2"] = 1
        if pred_status == "yes":
            patient_info_dict[patientID, "pred", "visit2"] = 1
        if mtx_status == "yes":
            patient_info_dict[patientID, "mtx", "visit2"] = 1
    
print (len(patientID_list))
```

```python
def create_contigency_table(DMARD, patient_info_dict):
    #Order: visit1, visit2
    give_give = 0
    notgive_give = 0
    give_notgive = 0
    notgive_notgive =0 
    
    visit1_count = 0
    visit2_count = 0
    
    for patientID in patientID_list:
        
        try: 
            visit1_status = patient_info_dict[patientID, DMARD, "visit1"]
        except KeyError: 
            visit1_status = 0
        try: 
            visit2_status = patient_info_dict[patientID, DMARD, "visit2"]
        except KeyError: 
            visit2_status = 0
        
        if visit1_status == 1 and visit2_status == 1:
            give_give += 1
        if visit1_status == 0 and visit2_status == 1:
            notgive_give += 1
        if visit1_status == 1 and visit2_status == 0:
            give_notgive += 1
        if visit1_status == 0 and visit2_status == 0:
            notgive_notgive += 1
            
        if visit1_status == 1:
            visit1_count += 1
        if visit2_status == 1:
            visit2_count += 1
    print (DMARD)
    print ("COUNTS", visit1_count, visit2_count)
    print (give_give + notgive_give + give_notgive + notgive_notgive)
    print ("g_g, ng_g, g_ng, ng_ng")
    print (give_give, notgive_give, give_notgive, notgive_notgive)
```

```python
create_contigency_table("csDMARD", patient_info_dict)
create_contigency_table("bDMARD", patient_info_dict)
create_contigency_table("pred", patient_info_dict)
create_contigency_table("mtx", patient_info_dict)
```
