```r
library(lme4)
library(lmerTest)
library(effects)
library(stringr)
```

```r
improve_data_dir <- "../preprocessed_data/RA.plasma.metabolites.improve.tsv"
none_data_dir <- "../preprocessed_data/RA.plasma.metabolites.none.tsv"
improve_data_df <- read.csv(improve_data_dir,sep="\t",row.names=1)
none_data_df <- read.csv(none_data_dir,sep="\t",row.names=1)

improve_data_df <- t(improve_data_df)
improve_data_df <- as.data.frame(improve_data_df)

none_data_df <- t(none_data_df)
none_data_df <- as.data.frame(none_data_df)
```

```r
col_list <- colnames(improve_data_df)
chemID_list <- col_list[6:length(col_list)]
```

```r
output_txt = '../improvement_analysis/improve_patients.MLR.reproduce.tsv'
if (file.exists(output_txt)){
    file.remove(output_txt)
}

output_string <- paste("chemID\timprove_pvalue\tnone_pvalue\tcriteria\n")
cat(output_string, file=output_txt,append=TRUE)

for (i in 1:length(col_list)){
    if (i > 4){
        chemID <- chemID_list[i-4]
        output_string <- chemID
        improve_lm <- glmer(timepoint ~ improve_data_df[,i] + SEX + AGE + (1|patient_dummyID), data= improve_data_df, family = binomial)
        improve_pvalue = coef(summary(improve_lm))[,4][2]

        none_lm <- glmer(timepoint ~ none_data_df[,i] + SEX + AGE + (1|patient_dummyID), data= none_data_df, family = binomial)
        none_pvalue <- coef(summary(none_lm))[,4][2]
        
        #eular diagram numbers 
        count = 0
        if (improve_pvalue < 0.05){
        count = count + 1
        }
        if (none_pvalue < 0.05){
        count = count + 2
        }

        output_string <- paste(output_string, '\t', improve_pvalue,'\t', none_pvalue,'\t',count,'\n', sep="")
        cat(output_string, file=output_txt,append=TRUE)
    }
    
}
```
