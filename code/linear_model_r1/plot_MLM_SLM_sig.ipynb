```r
library("ggplot2")
```

```r

data_file <- '/Users/m221138/RA_project/analysis/linear_model/plots/clp.mlm.sig.csv'
data_profile <- read.csv(data_file, sep="\t", header=TRUE)
```

```r
NUM_CHEM_ID <- ncol(data_profile)
NUM_CHEM_ID

chem_profile <- data_profile[,4]
DAS <- data_profile[,1]
SEX <- data_profile[,2]
AGE <- data_profile[,3]

MLM <- lm(DAS ~ chem_profile + SEX + AGE, data=data_profile)
SLM <- lm(DAS ~ chem_profile, data=data_profile)
# plot(SLM, which=1)
# plot(SLM, which=1)

# str(MLM)

ggplot(MLM, aes(x = .fitted, y = .resid)) + 
    geom_point(col="red") + 
    geom_point(aes(SLM$fitted,y =SLM$resid), col="blue") +
    geom_hline(yintercept=0, col="red", linetype="dashed") +
    stat_smooth(method="loess")

# for (i in 1:NUM_CHEM_ID){
#     if (i > 3){
#         output_string <- ''
        
#         das <- data_profile[,1]
#         chem_profile <- data_profile[,i]

#         SLM <- lm(DAS ~ chem_profile, data=data_profile)
#         MLM <- lm(DAS ~ chem_profile + SEX + AGE, data=data_profile)
#     }
# }
```
