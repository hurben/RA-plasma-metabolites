```r
input_data <- '/Users/m221138/RA_project/analysis/all_in_one_r2/hd4_only/full.ml.ready.txt'
input_df <- read.csv(input_data, header=TRUE, row.names=1, sep="\t")

NUM_ROW = nrow(input_df)
NUM_COL = ncol(input_df)

das28_list <- input_df[1,]


for (i in 1:NUM_COL){
    das28 = das28_list[i]
    if (das28 <= 3.2){
        category = 'low'
    }
    if (das28 > 3.2 & das28 <= 5.1){
        category = 'med'
    }
    if (das28 > 5.1 ){
        category = 'high'
    }
    das28_list[i] <- category
}
das28_meta <- data.frame(t(das28_list))

input_df <- as.matrix(input_df[2:NUM_ROW,])
col_names <- colnames(input_df)
row_names <- rownames(input_df)

rank_df <- rowRanks(-input_df)
colnames(rank_df) <- col_names
rownames(rank_df) <- row_names

pal <- colorRampPalette(c("yellow","red"))
pal <- pal(128)

#print (rank_df)
#heatmap_results <- pheatmap(rank_df, color=pal)
# sorted_patientID_list <- colnames(rank_df[,heatmap_results$tree_col[["order"]]])
# sorted_das28_list <- as.matrix(das28_list[,heatmap_results$tree_col[["order"]]])
# das_pal <- colorRampPalette(c("blue","yellow","red"))
# das_pal <- das_pal(128)

heatmap_results <- pheatmap(rank_df, color=pal, annotation_col = das28_meta)
sorted_patientID_list <- colnames(rank_df[,heatmap_results$tree_col[["order"]]])
sorted_das28_list <- as.matrix(das28_list[,heatmap_results$tree_col[["order"]]])

print (sorted_patientID_list)
```

```r
# make test matrix
test = matrix(rnorm(200), 20, 10)
test[1:10, seq(1, 10, 2)] = test[1:10, seq(1, 10, 2)] + 3
test[11:20, seq(2, 10, 2)] = test[11:20, seq(2, 10, 2)] + 2
test[15:20, seq(2, 10, 2)] = test[15:20, seq(2, 10, 2)] + 4
colnames(test) = paste("Test", 1:10, sep = "")
rownames(test) = paste("Gene", 1:20, sep = "")

# define the annotation
annotation_row = data.frame(
                    GeneClass = factor(rep(c("Path1", "Path2", "Path3"), c(10, 4, 6))),
                  AdditionalAnnotation = c(rep("random1", 10), rep("random2", 10))
                )
rownames(annotation_row) = paste("Gene", 1:20, sep = "")

pheatmap(test, annotation_row = annotation_row)
```

```r
data <- replicate(20, rnorm(50))
rownames(data) <- paste("Gene", c(1:nrow(data)))
colnames(data) <- paste("Sample", c(1:ncol(data)))
```

```r
metadata <- data.frame(
      c(rep("case", ncol(data)/2), rep("control", ncol(data)/2)),
      c(rep("cond1", ncol(data)/4), rep("cond2", ncol(data)/4), rep("cond3", ncol(data)/4), rep("cond4", ncol(data)/4)),
      row.names=colnames(data))
colnames(metadata) <- c("casecontrol","condition")

metadata_gene <- data.frame(
      c(rep("Tcell", nrow(data)/2), rep("Bcell", nrow(data)/2)),
      row.names=rownames(data))
colnames(metadata_gene) <- c("Cell")

out <- pheatmap(data, 
      show_rownames=F, cluster_cols=T, cluster_rows=T, scale="row",
      cex=1, clustering_distance_rows="euclidean", cex=1,
      clustering_distance_cols="euclidean", clustering_method="complete", border_color=FALSE,
      annotation_col=metadata,
      annotation_row=metadata_gene)
```
